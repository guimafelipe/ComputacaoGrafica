```python
from skimage import data, io
from matplotlib import pyplot as plt
```

```python
camera = data.camera()
```

```python
io.imshow(camera)
plt.show()
```

```python
# Acrescenta rudo gaussiano com 2 varincias
from skimage import util
camera_gaussianNoise1 = util.random_noise(camera, mode='gaussian', var=0.005)
camera_gaussianNoise2 = util.random_noise(camera_gaussianNoise1, mode='gaussian', var=0.003)
```

```python
io.imshow(camera_gaussianNoise2)
plt.show()
```

```python
# Acrescenta rudos salt-and-pepper com 2 valores de densidade
camera_saltPepperNoise1 = util.random_noise(camera_gaussianNoise2, mode='s&p', amount=0.001, salt_vs_pepper=0.5)
camera_saltPepperNoise2 = util.random_noise(camera_saltPepperNoise1, mode='s&p', amount=0.002, salt_vs_pepper=0.3)
```

```python
io.imshow(camera_saltPepperNoise2)
plt.show()
```

```python
# Filtro de mediana
from skimage import filters
camera_medianFilter1 = filters.median(camera_saltPepperNoise2)
camera_medianFilter2 = filters.median(camera_medianFilter1)
```

```python
io.imshow(camera_medianFilter2)
plt.show()
```

```python
# Filtro gaussiano com kernels de 2 tamanhos diferentes
camera_gaussianFilter1 = filters.gaussian(camera_medianFilter2, sigma=0.2)
camera_gaussianFilter2 = filters.gaussian(camera_gaussianFilter1, sigma=0.3)
```

```python
io.imshow(camera_gaussianFilter2)
plt.show()
```
